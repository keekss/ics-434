#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

disciplines = ['CS', 'Engineering', 'Math', 'Physics']

np.random.seed(42)

# Use indices of disciplines with np.random.randint with list
students_affiliations = np.array([disciplines[np.random.randint(4)] for i in range(1000)])
disciplines, counts = np.unique(students_affiliations, return_counts=True)
# argsort with descending order
indices_sorted = np.argsort(-counts)

# Use argsort indices to sort
disciplines_sorted = [disciplines[i] for i in indices_sorted]
counts_sorted = [counts[i] for i in indices_sorted]

# Add custom aspects to plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of Students in the Program', color = 'green')
ax.set_ylabel('Counts')
ax.set_xlabel('Disciplines')

ax.bar(disciplines_sorted, counts_sorted, color = "#FF5733", alpha = 0.55)
plt.show()
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
def random_scatter_squares(n):
    # Assign random values for x and y
    x, y, colors = np.random.rand(n), np.random.rand(n), np.random.rand(n)
    plt.scatter(x, y, c = colors, marker = 's')
    plt.show()
```

```python
random_scatter_squares(10)
```

```python
random_scatter_squares(100)
```

```python
random_scatter_squares(1000)
```
